```python
#import libraries
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data
uber_data = pd.read_csv('C:/Users/Saidul/Desktop/Uber Data Analysis with Python/uber-raw-data-apr14.csv')
uber_data.head()
```

```python
uber_data.info()
```

#convert the date time and add some new columns

```python
#convert date/time column data type
uber_data ['Date/Time'] = uber_data['Date/Time'].map(pd.to_datetime)
uber_data.tail()
```

```python
#add day column
def get_day_of_month(dt):
    return dt.day

uber_data['Day'] = uber_data['Date/Time'].map(get_day_of_month)
uber_data.tail()
```

```python
#add weekday and hour column

def get_weekday(dt):
    return dt.weekday()
uber_data['Week Day'] = uber_data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour
uber_data['Hour'] = uber_data['Date/Time'].map(get_hour)

uber_data.tail()
```

#analyse the data

```python
#plot the day data in a histogram
hist(uber_data.Day, bins=30, rwidth=.8, range=(0.5, 30.5))
xlabel('Date of The Month')
xticks(range(1,31), rotation ='vertical')
ylabel('Frequency')
title('Frequency vs Day')
```

```python
#print the frequency by a day
def count_rows(rows):
    return len(rows)

by_date = uber_data.groupby('Day').apply(count_rows)
by_date
```

```python
#line graph of the day
plot(by_date)
```

```python
#bar graph of the day
bar(range(1,31),by_date)
```

```python
#print the sorted value
sorted_day = by_date.sort_values()
sorted_day
```

```python
#plot the sorted value
bar(range(1, 31), sorted_day)
xticks(range(1,31), sorted_day.index, rotation = 'vertical')
xlabel('Date of The Month')
ylabel('Rrequency')
title('Frequency vs Day')
```

#cross analysis (Hour, DOW)

```python
by_cross = uber_data.groupby('Week_day Hour'.split()).apply(count_rows).unstack()
sns.heatmap(by_cross)
```
